```python
 class PlaceIDFilter:
    def __init__(self, df, keys=["BUILDINGID", "FLOOR", "SPACEID"]):
        self.number = 0
        self.df = df
        self.keys = keys
    def __call__(self, index):
        values = []
        for col in self.keys:
            values.append(str(self.df[col][index]))
        return "-".join(values)
```

```python
train_df_classification_dict = {a: b for a, b in train_df.groupby("PHONEID")}
# 为每个切片按时间排序
for df in train_df_classification_dict.values():
    df.sort_values("TIMESTAMP", inplace=True)
# 将每个切片按标签再次切片
items = train_df_classification_dict.items()
train_df_classification_dict = {}
for key, df in items:
    d = train_df_classification_dict[key] = {}
    pif = PlaceIDFilter(df)
    for key, df in df.groupby(pif):
        d.setdefault(key, []).append(df)
```
